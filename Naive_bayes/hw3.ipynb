```python
data = pd.read_csv('challengeTrainLabeled.csv')

def process(data):
    
    ## dropping data 
    data= data.drop(['patient_id','Status','Unnamed: 0','Release_Cycle','Race1','Q5','Q6','Q7','Q8',
                'fluoride_plasma_level','fluoride_water_level','Interview_lang','Self_respond_fam',
                'Income_family','Self_respond2_fam','HH_education','HH_country','HH_marital','HH_gender',
                'HH_age','weight_status','last_alcohol','pregnancy_test','dentition_status',
                'Self_respond_sp','Self_respond2_sp','Q9','Q11','Q13','Q19'],axis=1)
    
    data["Q27"] = pd.to_numeric(data.Q27, errors='coerce')
    
    ## filling the categorical data 
    
    data = data.fillna({'Q2':7,'Q3':7,'Q4':7,'Q12':7,'Q14':7,'Q15':7,'Q16':7,'Q18':7,'Q20':7,'Q21':7,'Q22':7,'Q26' : 7, 'Q27':7})
    data = data.fillna({'virusB_infect_status':3,'virusC_infect_status':3,'virusD_infect_status':3,'virusE_infect_status':3})
    data = data.fillna({'hepatitis_A_test':3, 'hepatitis_D_test' :3, 'hepatitis_B_test' :3})
#     data.to_csv('test2.csv')
    data['US_citizen'].fillna(1,inplace =True)
     

    data['Income_household'].fillna(np.mean(data['Income_household']),inplace =True)
    
    data['weight'].fillna(np.mean(data['weight']),inplace=True)
#     data.to_csv('test3.csv')
    
    # observing from age 0 -7 the imputing values : 
    for i in range(0,6):
#         print(i)
        blood_pressure=np.random.randint(50,65, size = (len(data['blood_pressure'].loc[data['Age']==i]),))
        data['blood_pressure'].loc[data['Age']==i]=blood_pressure

    # imputing values for max_level for age 0-7:  
    for i in range(0,6):
        max_level=np.random.randint(85,100, size = (len(data['blood_pressure'].loc[data['Age']==i]),))
        data['max_inflation_level'].loc[data['Age']==i]=max_level
        
    for i in range(0,6):
        max_level=np.random.randint(85,100, size = (len(data['blood_pressure'].loc[data['Age']==i]),))
        data['max_inflation_level'].loc[data['Age']==i]=max_level
        
    # for other ages
    data['max_inflation_level'].fillna(-5, inplace = True)
    data['blood_pressure'].fillna(-5, inplace=True)
    data['pulse_status'].fillna(-3,inplace=True)
    data['calcium_level'].fillna(-10,inplace=True)
    data['iron_level'].fillna(-10,inplace=True)
    data['Income_ratio'].fillna(np.mean(data['Income_ratio']),inplace=True)
    data['Q1'].fillna(7,inplace=True)
    data['Q10'].fillna(7,inplace=True)
    data['Q17'].fillna(9,inplace=True)
    data['Q23'].fillna(9,inplace=True)
    data['Q24'].fillna(9,inplace=True)
    data['Q25'].fillna(9,inplace=True)
    data['abdomen_diameter'].fillna(np.mean(data['abdomen_diameter']),inplace=True)
    data['glycohemoglobin'].fillna(-5, inplace=True)
    data[ 'recent_alcohol'].fillna(-5 , inplace = True)
    data[ 'glucose_level'].fillna(-5,inplace=True)
    #for range 
    
    for i in range(0,6):
        print(i)
        data['urine_flow_rate'].loc[data['Age']==i].fillna(-5,inplace=True)
     
    #data.to_csv('age,urine.csv')
    data['urine_flow_rate'].fillna(-10,inplace=True)
    
    
    return data 

# data.drop(['pregnancy_test','Unnamed: 0','HH_gender','Race2','Interview_lang','Self_respond_fam','Self_respond2_fam','Num_adolescent','Num_family','HH_marital','HH_country','Num_household','Num_adolescent','Income_ratio' ,'Num_young_child','Income_household','Income_family' ,'Num_elder' ,'HH_age' ,'HH_education','patient_id', 'Release_Cycle','Status','Country','Language','last_alcohol',
#            'Q1','Q5','Q6','Q7','Q8',
#            'Q9','Q11','Q10','Q13','Q17','Q19','Q23','dentition_status',
#            'Q24','Q25','Q25','weight_status','abdomen_diameter','fluoride_plasma_level','fluoride_water_level'],axis=1,inplace= True)

# data["Q27"] = pd.to_numeric(data.Q27, errors='coerce')

# data = data.fillna({'Q2':7,'Q3':7,'Q4':7,'Q12':7,'Q14':7,'Q15':7,'Q16':7,'Q18':7,'Q20':7,'Q21':7,'Q22':7,'Q26' : 7, 'Q27':7})
# data = data.fillna({'virusB_infect_status':3,'virusC_infect_status':3,'virusD_infect_status':3,'virusE_infect_status':3})
# data = data.fillna({'hepatitis_A_test':3, 'hepatitis_D_test' :3, 'hepatitis_B_test' :3})




data = process(data)

data.isna().sum()



# seaborn.heatmap(data,annot=True)
# for i in range(0,38):
#     print(i)
#     plt.hist(x=data.iloc[:,i])
#     plt.show()



# # use this for dimensionality relation 
# data.corr()


# imputing values for blood pressure for age 0 to 7  

    
        
    
    
    

    



# df=data.loc[data['max_inflation_level'].isnull()]
# df['Age'].value_counts()
#df.to_csv('check_age0.csv')


    
    
```

```python
import sklearn
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def svm_clf(data):
    scores=[]
    X = np.asarray(data.iloc[:,:-1])
    y = np.asarray(data.iloc[:,-1])
    
    # using the classifier 
    clf = svm.SVC(kernel='linear', C=1)
    scores = cross_val_score(clf, X, y, cv=5)
    
    return np.mean(scores)

def dec_clf(data):
    scores=[]
    X = np.asarray(data.iloc[:,:-1])
    y = np.asarray(data.iloc[:,-1])
    
    # usign the classfier 
    
    clf = DecisionTreeClassifier(criterion ='gini',random_state=0)
    scores = cross_val_score(clf, X, y, cv=5)
    return np.mean(scores)

def random_forest(data):
    scores=[]
    X = np.asarray(data.iloc[:,:-1])
    y = np.asarray(data.iloc[:,-1])
    
    clf=RandomForestClassifier(n_estimators=100, max_depth=5,random_state=0)
    
    scores = cross_val_score(clf, X,y,cv=5)
    
    return scores ,np.mean(scores)
    
def ada_clf(data):
    
    scores=[]
    scores1=[]
    X = np.asarray(data.iloc[:,:-1])
    y = np.asarray(data.iloc[:,-1])
        
    #clf =svm.SVC(gamma='auto',probability= True)
    
    clf = DecisionTreeClassifier(max_depth = 1,criterion ='gini',random_state=0)
    
    clf_ada = AdaBoostClassifier(n_estimators=200,base_estimator=clf,learning_rate=1)
   
    scores = cross_val_score(clf_ada, X,y,cv=5)
    
    return np.mean(scores) 
    
print(svm_clf(data))

#dec_clf(data)
print(ada_clf(data))

#random_forest(data)
```
